# Descriptive Statistics of data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
file_path = "/content/devdata.csv"
datacleaned_data = pd.read_csv(file_path)

# Explicitly convert inf values to NaN
datacleaned_data.replace([float('inf'), float('-inf')], pd.NA, inplace=True)

# Set a theme for the plots
sns.set_theme(style="whitegrid")

# 1. Basic Information about the dataset
print("Dataset Info:")
print(datacleaned_data.info())
print("\nFirst few rows:")
print(datacleaned_data.head())

# 2. Descriptive Statistics
print("\nDescriptive Statistics:")
print(datacleaned_data.describe())

# 3. Distribution of Population by Continent
plt.figure(figsize=(10, 6))
sns.boxplot(x='continent', y='population', data=datacleaned_data)
plt.title('Population Distribution by Continent', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Monthly Count Trend over time for entire world covid cases

```python
# Convert 'year_week' to a datetime format for easier manipulation
datacleaned_data['date'] = pd.to_datetime(datacleaned_data['year_week'] + '-1', format='%Y-%W-%w')

# Extract year and month for grouping
datacleaned_data['year_month'] = datacleaned_data['date'].dt.to_period('M')

# Group by 'year_month' and calculate the sum of 'weekly_count' for each month
monthly_data = datacleaned_data.groupby('year_month', as_index=False)['weekly_count'].sum()

# Convert 'year_month' to string for plotting compatibility
monthly_data['year_month'] = monthly_data['year_month'].astype(str)

# Create the plot with monthly aggregated data
plt.figure(figsize=(14, 6))
sns.lineplot(x='year_month', y='weekly_count', data=monthly_data, ci=None)
plt.title('Monthly Count Trend Over Time', fontsize=16)
plt.xlabel('Year-Month')
plt.ylabel('Monthly Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
# file_path = "/content/datacleaned_data.csv"
datacleaned_data = pd.read_csv(file_path)


top_countries = datacleaned_data.groupby('country')['weekly_count'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries by Total Weekly Counts', fontsize=16)
plt.xlabel('Total Weekly Count')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

# Show different numerical trends per population by time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
file_path = "/content/devdata.csv"  # Update with your actual file path
data = pd.read_csv(file_path)

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Specify columns to plot
numerical_cols = ['population', 'weekly_count', 'rate_14_day', 'cumulative_count']
existing_cols = [col for col in numerical_cols if col in numeric_data.columns]

# Create a time-like index if not available
if 'time' not in data.columns:
    data['time'] = range(len(data))  # Sequential index for plotting

# Plot line charts
plt.figure(figsize=(12, 6))
for col in existing_cols:
    plt.plot(data['time'], numeric_data[col], label=col, alpha=0.8, linewidth=2)

plt.title(" Trends of Numerical Features", fontsize=16)
plt.xlabel("Time Index")
plt.ylabel("Values")
plt.legend(title="Features")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
